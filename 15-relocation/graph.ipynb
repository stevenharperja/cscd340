```python
import subprocess
print(subprocess.run(['ls'], capture_output=True, text=True).stdout)

add_size = 1024
num_add = 100
avg_counts = []
for limit in range(1, add_size + 1):
    counts = []
    for seed in range(5):
        output = subprocess.run(['python2 relocation.py -c -s {seed} -n {num_add} -l {limit}'.format(seed=seed,num_add=num_add,limit=limit)], capture_output=True, text=True, shell=True).stdout

        #proc = subprocess.Popen(['python2 relocation.py -c -s {seed} -n {num_add} -l {limit}'.format('')], stdout=subprocess.PIPE, shell=True)
        #(out, err) = proc.communicate()
        #print("program output:", out)
        #print(output)
        count = output.count('VALID')
        counts.append(count)
    avg_counts.append(sum(counts) / len(counts))
```

```python
#graph the counts using matplotlib
import matplotlib.pyplot as plt
plt.plot(range(1, add_size + 1), avg_counts)
plt.xlabel('limit')
plt.ylabel('average valid')
plt.title('limit vs average valid')
plt.show()
```
